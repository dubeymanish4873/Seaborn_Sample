```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("https://raw.githubusercontent.com/Premalatha-success/Yhills_July12_Analytics/main/titanic-training-data.csv")
```

```python
df.head(20)
```

```python
sns.lineplot(x="Survived",y="Pclass",data=df,color="black",hue="Sex",style="Sex",palette="Accent"
             ,markers=['o','*'],dashes=False,legend="full")
plt.grid()
## In seaborn automatically the label option come on to the graph but in matplotlib we have use legend for it first.
## Linear plot
```

```python
sns.lineplot(x="Name",y="Sex",data=df,hue="Age")
```

```python
df.head(20)
```

```python
sns.catplot(x="Sex",y="Age",jitter=False,data=df)
## Jitter Plot
```

```python
sns.catplot(x="Sex",y="Age",hue="Pclass",data=df)
## Hue Plot
```

```python
sns.relplot(x="PassengerId",y="Age",hue="Sex",data=df)
```

```python
sns.catplot(x="PassengerId",y="Age",kind="swarm",data=df)
```

```python
sns.catplot(x="Sex",y="Age",kind="swarm",data=df)
## Swarm Plot
```

```python
sns.catplot(x="Sex",y="Age",kind="swarm",hue="Pclass",data=df)
```

```python
## Box-Plot
sns.catplot(x="Sex",y="Age",kind="box",data=df)
```

```python
sns.catplot(x="Sex",y="Age",kind="box",hue="Pclass",data=df)
```

```python
## Violin Plot
sns.catplot(x="Sex",y="Age",kind="violin",data=df)
```

```python
sns.catplot(x="Sex",y="Age",kind="violin",hue="Pclass",marker=['*',"o"],data=df)
```

```python
##  BoxPlot
sns.catplot(x="Sex",y="Age",kind="bar",data=df)
```

```python
sns.catplot(x="Sex",y="Age",kind="bar",hue="Pclass",data=df)
```
